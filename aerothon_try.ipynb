```python
import cv2 as cv
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

def classify_shape(approx):
    if len(approx) == 3:
        return "Triangle"
    elif len(approx) == 4:
        (x, y, w, h) = cv.boundingRect(approx)
        ar = w / float(h)     #aspect ratio, ratio of width and height
        return "Square" if 0.95 <= ar <= 1.05 else "Rectangle"
    elif len(approx) == 5:
        return "Pentagon"
    elif 6 < len(approx) < 15:
        return "Ellipse"
    else:
        return "Circle"

uploaded = files.upload()
image_name = next(iter(uploaded))

image = cv.imread(image_name)
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray, (5, 5), 0)
edged = cv.Canny(blurred, 50, 150)

contours, _ = cv.findContours(edged.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
#chain approx simple reduces the no of vertices to min no of points reqd to define that shape
#external makes us take only external contours
#edged copy means you won't edit original image
shape_count = {}

for contour in contours:
    approx = cv.approxPolyDP(contour, 0.04 * cv.arcLength(contour, True), True)
    shape = classify_shape(approx)

    # Updating the shape count
    if shape in shape_count:
        shape_count[shape] += 1
    else:
        shape_count[shape] = 1

    cv.drawContours(image, [approx], -1, (0, 255, 0), 2)
    cv.putText(image, shape, (approx.ravel()[0], approx.ravel()[1]), cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

# Display the processed image with detected shapes
cv2_imshow(image)

# Display shape counts
print("Shape Counts:")
for shape, count in shape_count.items():
    print(f"{shape}: {count}")
```
